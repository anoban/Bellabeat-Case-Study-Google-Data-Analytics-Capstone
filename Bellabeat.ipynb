### **Data source:**
The data needed for this case study is hosted in **Kaggle** https://www.kaggle.com/datasets/arashnic/fitbit published by a Kaggle user named **`Mobius`**.    
Datasets were generated through a distributed survey via Amazon Mechanical Turk between **`03.12.2016` - `05.12.2016`**.   
This Kaggle data set contains personal ﬁtness tracker from thirty Fitbit users. Thirty eligible Fitbit users consented to the submission of personal tracker data, including minute-level output for physical activity, heart rate, and sleep monitoring. It includes information about daily activity, steps, and heart rate that can be used to explore users’ habits.

### **Limitations:**
1. The datasets are rather poorly organized, instead of making a single dataset with the lowest possible timescale i.e minutes there are a bloated number of datasets that summarize same parameters on different timescales i.e minutes wise, hourly and daily summaries. This makes analytics cumbersome.     
2. The variable names (Headers) of the .csv files are not properly documented. One cannot take a header for representing a feature for sure.        
3. The datasets have no licesnces explicitly stated. There is no way to be sure whether these datasets are allowed for public usage.      
4. Since these files are hosted on a generic data analysis related social platform the credibility of the data becomes a question mark!        
5. And these datasets have records from just 30 participants who consented for the data collection. 30 is an extremely small number to be representative of all the Fitbit users. It is very likely this will result in unacceptable sampling errors. One cannot make business decisions based on records from just 30 users.     
6. Besides these data are from a totally different smart watch company named Fitbit, and we do not know how the insights gained from such a large scale global gadget manufacturer will translate to the context of Bellabeat.     
7. Since Bellabeat strictly focuses on smart health gadgets for women, we do not know for certain that usage records from 30 random users whose genders are unbeknownst to us can be the best data to analyze in order to design business strategies.        

### **ROCCC:**
1. Reliable - `2/10` No explicit details of data gathering process. How did they harvest records over many months of usage through a "survey" that lasted just 2 days? Have there any cleaning/preprocessing happened that might have altered the data?       
2. Original - `1/10` Since the data is allegedly claimed to have come from Fitbit smart watches no information about the gadget models, the sensors & meters installed in those devices used to gather such measurements are explicitly cited.          
3. Comprehensive - `1/10` The archive contains unnecessarily bloated number of datasets with filenames poorly informative of the data they contain. Headers of the datasets are hard to understand at certain instances. A separate text or markdown file documenting the headers and files could have been very helpful.          
4. Current - `1/10` The survey conducted to gather these data happened in December 2016, roughly six years before. technology has evolved much in the past six years and user preferences & usage patterns have changed drastically. Thus these datasets are very outdated to make any serious marketing decisions.         
5. Cited - `0/10` No credible academic citations were found for these Fitbit tracker datasets. The only citations found were from the capstone projects by Google which also makes the credibility of these datasets highly questionable.         

**Due to all the above mentioned shortcomings these datasets are a very poor choice for data analytics**

```r
hrate_8792009665 <- hrate %>% filter(Id == "8792009665")
isect_8792009665 <- narrowed_isect %>% filter(Id == "8792009665")
```

```r
time_delta <- strptime("14-03-2022 01:00:00", format = "%d-%m-%Y %H:%M:%S") - strptime("13-03-2022 24:00:00", format = "%d-%m-%Y %H:%M:%S")
```

```r
hrate_8792009665 %>% head()
```

```r
hourly_mean_hrate <- c()  # initializing an empty vector to append the averages of hourly hearbeats
hbeat <- c()  # initializing an empty vector to append heart rates within the span of an hour
start_ind = 1  # end index of every hour in the timeseries, will be updated via the for loop

for (i in 1:dim(hrate_8792009665)[1]) {   # looping over all rows
        if (is.na(hrate_8792009665$Time[i] - hrate_8792009665$Time[start_ind])) {
            next }  # skip to next the next value in the iterable if the subtraction results in an NA
        else if ((hrate_8792009665$Time[i] - hrate_8792009665$Time[start_ind]) <= time_delta) { # if the subtraction gives a value less than an hour
            hbeat <- c(hbeat, hrate_8792009665$Value[i]) }  # add the heart beat value to the vector hbeat
        else {  # if the subtraction gives a value larger than an hour
            hourly_mean_hrate <- c(hourly_mean_hrate, mean(hbeat, na.rm = TRUE))  
            # means that we have surpassed the end limit of an hour, compute the mean of all heart beats within that hour
            # append the computed hourly mean to the vector hourly_mean_hrate
            start_ind <- i  # update the start_ind such that the start limit for subtarction in the next interation
            # is the subtraction end limit of the current iteration
            hbeat <- c() # empty the vector, in preparation for next iteration 
            }
        }
```

In the case of `VeryActiveMinutes` and `FairlyActiveMinutes`, most data points are limited to 100 minutes. And the distribution of calories burnt fall between 1000 to 4000 with an obvious increase in burnt calories with minutes.

One meaningful inference that can be made from this is that users remain lightly active for more time compared to very or fairly active times. Perhaps because one naturally becomes exhausted after a short time of aggressive physical activity. But a striking observation is that **they burn more calories overall, in the lightly active minutes!**.  

And there are very few zeros in `lightlyActiveMinutes` & `SedentaryMinutes` compared to `VeryActiveMinutes` & `FairlyActiveMinutes` that have a dense vertical line of data points where x = 0!. This implies that there are **more poeple engaged in light physical activity compared to heavy physical labour or people spend more time in light physical sctivity over heavy activities** and it appears to work for them!.    

In the case of `SedentaryMinutes`, there appears to be two distinct clusters. Obviously people spend more minutes idle. And they burn quite similar amount of calories even when they are idle!.    

So to conclude, people spend more time being idle, followed by light activity and lesser times being fairly & highly active. And the type of activity appears to have little influence in the amonut of calories burnt!
